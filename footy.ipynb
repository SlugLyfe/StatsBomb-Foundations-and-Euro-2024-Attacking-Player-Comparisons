```python
def adjust_event_data(events_df): #adjusts pass and cary start and end locations. Adds succesful final third pass column.
    # separate start and end locations from coordinates
    events_df[['x', 'y']] = events_df['location'].apply(pd.Series)
    events_df[['pass_end_x', 'pass_end_y']] = events_df['pass_end_location'].apply(pd.Series)
    events_df[['carry_end_x', 'carry_end_y']] = events_df['carry_end_location'].apply(pd.Series)
    # adding section to events data that specifies if the pass was a succesful final third pass
    for index, row in events_df.iterrows():
        if (row['type'] == 'Pass') & (row['x'] < 80) & (row['pass_end_x'] > 80) & (pd.isna(row['pass_outcome'])):
            events_df.at[index, 'final_third_pass_complete'] = True
        else:
            events_df.at[index, 'final_third_pass_complete'] = False
    return events_df
     
```

```python
# A function which creates a dataframe with player names and the number of shots they have taken, xG, and goals scored. It takes in the events dataframe and a dataframe containing player names.
def shot_count_df_creator(events_df, player_list):
    # Filter the DataFrame to include only shots by players in the player list
    shots_df = events_df[(events_df['type'] == 'Shot') & events_df['player'].isin(player_list['player_name']) & (events_df['period'] != 5)]
    # Count the number of shots by player
    shot_counts_df = shots_df['player'].value_counts().reset_index()
    shot_counts_df['xG'] = 0
    shot_counts_df['goals'] = 0
    shot_counts_df.columns = ['player_name', 'shot_count', 'xG', 'goals']
    for index, row in shots_df.iterrows():
        name_of_player = row['player']
        shot_counts_df.loc[shot_counts_df['player_name'] == name_of_player, 'xG'] += row['shot_statsbomb_xg']

        if row['shot_outcome'] == 'Goal':
            shot_counts_df.loc[shot_counts_df['player_name'] == name_of_player, 'goals'] += 1
    return shot_counts_df
```

```python
# A function which creates a dataframe with player names and the number of passes succesfull passes they have made, their pass accuracy xA, assists, key passes, final third passes, through balls, and final third receptions. It takes in the events dataframe and a dataframe containing player names.
def pass_count_df_creator(events_df, player_list):
    passes_df = events_df[(events_df['type'] == 'Pass') & events_df['player'].isin(player_list['player_name'])]
    pass_counts_df = passes_df['player'].value_counts().reset_index()
    pass_counts_df['xA'] = 0
    pass_counts_df['assists'] = 0
    pass_counts_df['final third passes'] = 0
    pass_counts_df['through balls'] = 0
    pass_counts_df['key passes'] = 0
    pass_counts_df['succesful_pass_count'] = 0
    pass_counts_df['pass_accuracy'] = 0
    pass_counts_df['final_third_receptions'] = 0
    pass_counts_df.columns = ['player_name', 'attempted_pass_count', 'succesful_pass_count', 'pass_accuracy', 'xA', 'assists', 'key passes', 'final third passes', 'through balls', 'final_third_receptions']
    for index, row in passes_df.iterrows():
        name_of_player = row['player']
        name_of_receiver = row['pass_recipient']
        if row['pass_goal_assist'] == True:
            pass_counts_df.loc[pass_counts_df['player_name'] == name_of_player, 'assists'] += 1
        if row['pass_shot_assist'] == True:
            pass_counts_df.loc[pass_counts_df['player_name'] == name_of_player, 'key passes'] += 1
            pass_counts_df.loc[pass_counts_df['player_name'] == name_of_player, 'xA'] += xA_finder(row['pass_assisted_shot_id'],events_df)
        if row['final_third_pass_complete'] == True:
            pass_counts_df.loc[pass_counts_df['player_name'] == name_of_player, 'final third passes'] += 1
            if player_list['player_name'].str.contains(name_of_receiver).any():
                pass_counts_df.loc[pass_counts_df['player_name'] == name_of_receiver, 'final_third_receptions'] += 1
        if pd.isna(row['pass_outcome']):
            pass_counts_df.loc[pass_counts_df['player_name'] == name_of_player, 'succesful_pass_count'] += 1 
            if row['pass_through_ball'] == True:
                pass_counts_df.loc[pass_counts_df['player_name'] == name_of_player, 'through balls'] += 1  
    pass_counts_df['pass_accuracy'] = pass_counts_df['succesful_pass_count'] / pass_counts_df['attempted_pass_count'] * 100
    pass_counts_df = pass_counts_df.drop('attempted_pass_count', axis=1)
    return pass_counts_df
```

```python
# A function which takes in the pass_assisted_shot_id and the events dataframe, and returns the xG of the shot associated with that pass.
def xA_finder(pass_assisted_shot_id, events_df):
    # Find the row in events_df where the shot_id matches pass_assisted_shot_id
    shot_row = events_df[events_df['id'] == pass_assisted_shot_id]
    if not shot_row.empty:
        return shot_row.iloc[0]['shot_statsbomb_xg']

   
```

```python
# A function which creates a dataframe with player names and the number of dribbles they have completed, their dribble success rate, and the number of carries into the final third. It takes in the events dataframe and a dataframe containing player names.
def dribble_count_df_creator(events_df, player_list):
    dribble_df = events_df[((events_df['type'] == 'Dribble') | (events_df['type'] == 'Carry')) & events_df['player'].isin(player_list['player_name'])]
    # Count the number of shots by player
    dribble_counts_df = dribble_df['player'].value_counts().reset_index()
    dribble_counts_df['succesful dribbles'] = 0
    dribble_counts_df['final third carries'] = 0
    dribble_counts_df['dribble success rate'] = 0
    dribble_counts_df['attempted dribbles'] = 0
    dribble_counts_df.columns = ['player_name', 'attempted dribbles and carries', 'attempted dribbles', 'succesful dribbles', 'dribble success rate', 'final third carries']
    for index, row in dribble_df.iterrows():
        name_of_player = row['player']
        if row['type'] == 'Dribble':
            dribble_counts_df.loc[dribble_counts_df['player_name'] == name_of_player, 'attempted dribbles'] += 1
            if row['dribble_outcome'] == 'Complete':
                dribble_counts_df.loc[dribble_counts_df['player_name'] == name_of_player, 'succesful dribbles'] += 1
        if row['type'] == 'Carry':
            if row['carry_end_x'] > 80 and row['x'] < 80:
                dribble_counts_df.loc[dribble_counts_df['player_name'] == name_of_player, 'final third carries'] += 1
    dribble_counts_df['dribble success rate'] = dribble_counts_df['succesful dribbles'] / dribble_counts_df['attempted dribbles'] * 100
    dribble_counts_df = dribble_counts_df.drop('attempted dribbles', axis=1)  
    dribble_counts_df = dribble_counts_df.drop('attempted dribbles and carries', axis=1)
    return dribble_counts_df
```

```python
# A function which creates a dataframe with player names and the number of succesfful: interceptions, tackles, interceptions + tackles, ball recoveries and duels they have completed, along with their duel success rate
def defensive_actions_df_creator(events_df, player_list):
        defensive_df = events_df[((events_df['type'] == 'Interception') | (events_df['type'] == 'Duel')) | (events_df['type'] == 'Pressure')| (events_df['type'] == 'Ball Recovery') & events_df['player'].isin(player_list['player_name'])]
        defensive_counts_df = defensive_df['player'].value_counts().reset_index()
        defensive_counts_df['successful interceptions'] = 0
        defensive_counts_df['successful tackles'] = 0
        defensive_counts_df['pressures'] = 0
        defensive_counts_df['successful ball recoveries'] = 0
        defensive_counts_df['successful duels'] = 0
        defensive_counts_df['duels attempted'] = 0
        defensive_counts_df['duel success rate'] = 0
        defensive_counts_df["successful tackles and interceptions"] = 0
        defensive_counts_df.columns = ['player_name', 'throw away variable', 'successful interceptions', 'successful tackles', 'pressures', 'successful ball recoveries', 'successful duels', 'duels attempted', 'duel success rate', 'successful tackles and interceptions']
        for index, row in defensive_df.iterrows():
                name_of_player = row['player']
                if row['type'] == 'Interception':
                        if row['interception_outcome'] == 'Won' or row['interception_outcome'] == 'Success In Play':
                                defensive_counts_df.loc[defensive_counts_df['player_name'] == name_of_player, 'successful interceptions'] += 1
                if row['type'] == 'Pressure':
                        defensive_counts_df.loc[defensive_counts_df['player_name'] == name_of_player, 'pressures'] += 1
                if row['type'] == 'Duel':
                        defensive_counts_df.loc[defensive_counts_df['player_name'] == name_of_player, 'duels attempted'] += 1
                        if row['duel_outcome'] == 'Won' or row['duel_outcome'] == 'Success In Play':
                                defensive_counts_df.loc[defensive_counts_df['player_name'] == name_of_player, 'successful duels'] += 1
                        if row['duel_type'] == 'Tackle':
                                if row['duel_outcome'] == 'Won' or row['duel_outcome'] == 'Success In Play':
                                        defensive_counts_df.loc[defensive_counts_df['player_name'] == name_of_player, 'successful tackles'] += 1
                if (row['type'] == 'Ball Recovery') and pd.isna(row['ball_recovery_recovery_failure']):
                        defensive_counts_df.loc[defensive_counts_df['player_name'] == name_of_player, 'successful ball recoveries'] += 1
        
        defensive_counts_df['duel success rate'] = defensive_counts_df['successful duels'] / defensive_counts_df['duels attempted'] * 100
        defensive_counts_df['successful tackles and interceptions'] = defensive_counts_df['successful tackles'] + defensive_counts_df['successful interceptions']
        defensive_counts_df = defensive_counts_df.drop('throw away variable', axis=1)  
        defensive_counts_df = defensive_counts_df.drop('duels attempted', axis=1)
        return defensive_counts_df



```

```python
#merging all 4 dataframes (and adding xGC and final third progressions)
def merge_count_df_creator(events_df, player_list):
    shots_df = shot_count_df_creator(events_df, player_list)
    passes_df = pass_count_df_creator(events_df, player_list)
    dribbles_df = dribble_count_df_creator(events_df, player_list)
    defensive_df = defensive_actions_df_creator(events_df, player_list)
    merged_df = pd.merge(shots_df, passes_df, on='player_name')
    merged_df = pd.merge(merged_df, dribbles_df, on='player_name')
    print("merged_df columns:", merged_df.columns)
    merged_df = pd.merge(merged_df, defensive_df, on='player_name')
    merged_df['xGC'] = merged_df['xG'] + merged_df['xA'] # Adding expected goal contributions
    merged_df['final_third_progressions'] = merged_df['final_third_receptions'] + merged_df['final third carries'] # Adding final third progressions
    return merged_df
```

```python
#returns dataframe with all players in the player list and the number of minutes they have played
def minutes_calculator(matches, player_list):
    all_times = pd.DataFrame()
    for MATCH_ID in matches['match_id']:
        lineups = sb.lineups(match_id=MATCH_ID)
        events = sb.events(match_id=MATCH_ID)
        game_period_list = events['period'].unique()
        # keep only substitution events
        subs = events[events['type'] == 'Substitution']
        outs = subs[['player', 'team', 'minute']].rename(columns={'player':'player_name_out',
                                                                    'team':'team_name_out',
                                                                    'minute':'minute_out'})
        ins = subs[['substitution_replacement', 'team', 'minute']].rename(columns={'substitution_replacement':'player_name_in',
                                                                            'team':'team_name_in',
                                                                            'minute':'minute_in'})
        times = pd.merge(outs, ins, 
                         left_on=['player_name_out','team_name_out'], 
                         right_on=['player_name_in','team_name_in'], 
                         how='outer')
        times['player_name'] = times['player_name_out'].fillna(times['player_name_in'])
        
        starting_xi_events = events[events['type'] == 'Starting XI']
        starters_df = pd.DataFrame()
        for XI, TeamT in zip(starting_xi_events['tactics'], starting_xi_events['team']):
            team_starters_df = pd.DataFrame({'player_name': [player_entry['player']['name'] for player_entry in XI['lineup']]})
            starters_df = pd.concat([starters_df, team_starters_df], ignore_index=True)
        time = pd.merge(times, starters_df)




        times = pd.merge(times, starters_df, on='player_name', how='outer')
        times = times[times['player_name'].isin(player_list['player_name'])]
        if 3 in game_period_list:
            ET = True
        else:
            ET = False
        times['minutes_played'] = times.apply(minutes_played, ET=ET, axis=1)  # Pass ET
        all_times = pd.concat([all_times, times])
    all_times['minutes_played'] = all_times['minutes_played'].fillna(0)
    #all_times['player_name'] = all_times['player_name_out'].fillna(all_times['player_name_in'])
    minutes_played_df = all_times.groupby('player_name')['minutes_played'].sum().reset_index()
    minutes_played_df['minutes_played'] = minutes_played_df['minutes_played'].apply(lambda x: max(0, x))
    return minutes_played_df

def minutes_played(row, ET): # row = a row of the all times data frame which contain the substitution events, players, and timings
    if ET == True:
        full_length = 120
    else:
        full_length = 90
    if pd.notna(row.minute_out):   # starter who got subbed off
        return row.minute_out
    if pd.notna(row.minute_in):    # bench sub who came on
        return full_length - row.minute_in
    # no in or out = starter who played whole game
    return full_length


       

```

```python
#Takes in dataframe and columns to exclude, and adds new collumns for every statistic relating to its percentile.
def convert_to_percentile(df, columns_to_exclude):
    number_of_rows = df.shape[0]
    print("number of rows:", number_of_rows)
    columns = df.columns.difference(columns_to_exclude).tolist()
    df_percentile = df.copy()
    df_percentile[columns] = df[columns].rank(pct=True) * 100
    return df_percentile
    
```

```python
# A function which takes in a player, a dataframe with percentiles, a dataframe with per 90 stats, a list of parameters to compare, and the number of players to return. It returns a dataframe with the players most similar to the player passed in, along with printing a radar comparison of the most similar player.
def find_similar_players(player_name, percentile_df, per_90_df, params, num_players):
    player_stats = percentile_df[percentile_df['player_name'] == player_name]
    player_stats = player_stats[params].values.flatten().tolist()
    player_stats = [float(x) for x in player_stats]
    player_stats = [round(x, 2) for x in player_stats]
    # Calculate the absolute differences between the target player's stats and all other players' stats
    differences = percentile_df[params].sub(player_stats).abs().sum(axis=1)
    # Get the indices of the closest players
    closest_indices = differences.nsmallest(num_players + 1).index.tolist()
    # Exclude the target player from the results
    closest_indices.remove(percentile_df[percentile_df['player_name'] == player_name].index[0])
    # Create a DataFrame with the closest players' stats
    closest_players_df = percentile_df.iloc[closest_indices].copy().reset_index(drop=True)
    # Dividing the difference by the number of stats to get a final difference score
    closest_players_df['difference_avg'] = differences[closest_indices].values / len(params)
    # Get the names of the closest players
    closest_players = percentile_df.loc[closest_indices, 'player_name'].tolist()
    print("The player(s) most similar to", player_name, "based on the specified stats are:")
    print(closest_players)

    create_radar_comparison(percentile_df, [player_name, closest_players[0]], params, per_90_df)

    return closest_players_df
```
